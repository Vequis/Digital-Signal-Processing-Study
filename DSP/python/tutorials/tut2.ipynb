```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
```

```python
def generate_h(N):
    t = np.arange(-N, N+1, dtype=float)
    h = np.where(t == 0, 0, (-1)**t / t)
    return h

plt.stem(generate_h(20))
```

```python
import math

def fat(n):
    return math.factorial(int(n))
```

```python
N = 100
t = np.arange(-N, N+1, dtype=float)

plt.figure(figsize=(10, 10))

gamma = 1
b = [(fat(gamma * N)**2)/((fat(gamma * N - i)) * (fat(gamma * N + i))) for i in t]
plt.subplot(311)
plt.stem(b)
plt.title("gamma = 1")

gamma = 5
b = [(fat(gamma * N)**2)/((fat(gamma * N - i)) * (fat(gamma * N + i))) for i in t]
plt.subplot(312)
plt.stem(b)
plt.title("gamma = 5")

gamma = 100
b = [(fat(gamma * N)**2)/((fat(gamma * N - i)) * (fat(gamma * N + i))) for i in t]
plt.subplot(313)
plt.stem(b)
plt.title("gamma = 100")

plt.tight_layout()
plt.show()
```

```python
def generate_b(N, gamma):
    t = np.arange(-N, N+1, dtype=float)
    b = [(fat(gamma * N)**2)/((fat(gamma * N - i)) * (fat(gamma * N + i)) ) for i in t]
    return b

plt.stem(generate_b(20, 1))
```

```python
gamma_values =  [1, 5, 100]

plt.figure(figsize=(6, 6))

for i, gamma in enumerate(gamma_values):
    plt.plot(generate_b(20, gamma))
    plt.title(f"gamma = {gamma}")


plt.grid()
plt.legend()
```

```python
gamma_values = [1, 5, 100]
N_values = [20]

plt.figure(figsize=(6, 6))

for i, N in enumerate(N_values):
    for j, gamma in enumerate(gamma_values):

        h1 = generate_h(N)* generate_b(N, gamma)
        w, H = signal.freqz(h1, worN=4096)


        # plt.subplot(len(N_values), len(gamma_values), i*len(gamma_values) + j + 1)
        plt.plot(w, np.abs(H))
        plt.title("N = {}, gamma = {}".format(N, gamma))
```
