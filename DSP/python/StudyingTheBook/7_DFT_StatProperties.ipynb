```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = 1 * np.random.randn(64)

plt.stem(x)
plt.show()

print(np.mean(x))
print(np.var(x))
```

```python
def generate_data(n, mean=0, var=1):
    return np.random.randn(n) * np.sqrt(var) + mean
```

```python
# We're going to get the distribution of the n-th leg

def get_nth_leg(n, data):
    return (np.fft.fft(data))[n]

```

```python
real_results = []
complex_results = []

for i in range(50000):
    data = generate_data(64, mean = 0, var=4)
    real_results.append(np.real(get_nth_leg(1, data)))
    complex_results.append(np.imag(get_nth_leg(1, data)))

plt.hist(real_results, bins=100)
plt.hist(complex_results, bins=100)
plt.show()


print(np.mean(real_results), np.var(real_results))
print(np.mean(complex_results), np.var(complex_results))
```
